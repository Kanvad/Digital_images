1. Viết chương trình nạp một ảnh và lưu thành 3 ảnh với 3 màu khác nhau

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# Nạp ảnh gốc
try:
    original_image = iio.imread('bird.png')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'bird.png'. Vui lòng kiểm tra lại đường dẫn.")
    exit() # Thoát chương trình nếu không tìm thấy ảnh

# Kiểm tra xem ảnh có phải là ảnh màu (có 3 RGB) hay không
if original_image.ndim != 3 or original_image.shape[2] != 3:
    print("Lỗi: Ảnh nạp vào không phải là ảnh màu RGB hoặc có số không đúng.")
    exit()

# Tạo ảnh chỉ chứa màu Đỏ (Red)
# Tạo một mảng NumPy rỗng có cùng kích thước và kiểu dữ liệu với ảnh gốc
red_image = np.zeros_like(original_image)
red_image[:, :, 0] = original_image[:, :, 0] # Đỏ (index 0)

# Tạo ảnh chỉ chứa màu Xanh lá (Green)
green_image = np.zeros_like(original_image)
green_image[:, :, 1] = original_image[:, :, 1] # Xanh lá (index 1)

# Tạo ảnh chỉ chứa màu Xanh dương (Blue)
blue_image = np.zeros_like(original_image)
blue_image[:, :, 2] = original_image[:, :, 2] # Xanh dương (index 2)

# Lưu các ảnh mới
iio.imsave('bird_red.png', red_image)
iio.imsave('bird_green.png', green_image)
iio.imsave('bird_blue.png', blue_image)

print("Đã lưu các ảnh")

# Hiển thị các ảnh
plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(original_image)
plt.title('Ảnh gốc')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(red_image)
plt.title('Đỏ')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(green_image)
plt.title('Xanh lá')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(blue_image)
plt.title('Xanh dương')
plt.axis('off')

plt.tight_layout() # Điều chỉnh layout để các biểu đồ không chồng chéo
plt.show()
```

2. Viết chương trình nạp một ảnh và hoán đổi giá trị các màu. Lưu các ảnh vào máy

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# Nạp ảnh gốc
try:
    original_image = iio.imread('bird.png')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'bird.png'. Vui lòng kiểm tra lại đường dẫn.")
    exit() # Thoát chương trình nếu không tìm thấy ảnh

# Kiểm tra xem ảnh có phải là ảnh màu (có 3 RGB) hay không
if original_image.ndim != 3 or original_image.shape[2] != 3:
    print("Lỗi: Ảnh nạp vào không phải là ảnh màu RGB hoặc có số không đúng.")
    exit()

# Ảnh gốc thường có thứ tự kênh là Red (0), Green (1), Blue (2)

# Hoán đổi các kênh màu và lưu ảnh
# 1: Hoán đổi kênh Đỏ và Xanh dương (RGB -> BGR)
bgr_image = original_image[:, :, [2, 1, 0]]
# Lưu ảnh BGR
iio.imsave('bird_bgr_swapped.png', bgr_image)
print("Đã lưu bird_bgr_swapped.png (Hoán đổi Đỏ <-> Xanh dương)")

# 2: Hoán đổi kênh Đỏ và Xanh lá (RGB -> GRB)
grb_image = original_image[:, :, [1, 0, 2]]
# Lưu ảnh GRB
iio.imsave('bird_grb_swapped.png', grb_image)
print("Đã lưu bird_grb_swapped.png (Hoán đổi Đỏ <-> Xanh lá)")

# 3: Đảo ngược thứ tự kênh (RGB -> GBR)
gbr_image = original_image[:, :, [1, 2, 0]]
# Lưu ảnh GBR
iio.imsave('bird_gbr_swapped.png', gbr_image)
print("Đã lưu bird_gbr_swapped.png (Hoán đổi RGB -> GBR)")

# Hiển thị các ảnh
plt.figure(figsize=(18, 6))

plt.subplot(1, 4, 1)
plt.imshow(original_image)
plt.title('Ảnh Gốc (RGB)')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(bgr_image)
plt.title('Ảnh BGR (Đỏ <-> Xanh dương)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(grb_image)
plt.title('Ảnh GRB (Đỏ <-> Xanh lá)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(gbr_image)
plt.title('Ảnh GBR (RGB -> GBR)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

3. Viết chương trình nạp một ảnh, chuyển thành hệ màu HSV và lưu 3 ảnh với 3 màu khác nhau.

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
from skimage import color

# Nạp ảnh gốc
try:
    original_image_rgb = iio.imread('bird.png')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'bird.png'. Vui lòng kiểm tra lại đường dẫn.")
    exit() # Thoát chương trình nếu không tìm thấy ảnh

# Kiểm tra xem ảnh có phải là ảnh màu (có 3 RGB) hay không
if original_image_rgb.ndim != 3 or original_image_rgb.shape[2] != 3:
    print("Lỗi: Ảnh nạp vào không phải là ảnh màu RGB hoặc có số không đúng.")
    exit()

# Chuyển đổi ảnh từ hệ màu RGB sang HSV
hsv_image = color.rgb2hsv(original_image_rgb)

# Trích xuất từng kênh H, S, V
hue_channel = hsv_image[:, :, 0]       # Kênh Hue (Sắc thái màu)
saturation_channel = hsv_image[:, :, 1] # Kênh Saturation (Độ bão hòa màu)
value_channel = hsv_image[:, :, 2]     # Kênh Value (Độ sáng/Giá trị)

# Chuyển đổi các kênh về dạng uint8 [0-255] để lưu ảnh grayscale
hue_display = (hue_channel * 255).astype(np.uint8)
saturation_display = (saturation_channel * 255).astype(np.uint8)
value_display = (value_channel * 255).astype(np.uint8)

# Lưu 3 ảnh với 3 kênh khác nhau
iio.imsave('bird_hue_channel.png', hue_display)
iio.imsave('bird_saturation_channel.png', saturation_display)
iio.imsave('bird_value_channel.png', value_display)

print("Đã lưu các ảnh:")
print("- bird_hue_channel.png (Kênh Hue)")
print("- bird_saturation_channel.png (Kênh Saturation)")
print("- bird_value_channel.png (Kênh Value)")

# Hiển thị các ảnh
plt.figure(figsize=(16, 5))

plt.subplot(1, 4, 1)
plt.imshow(original_image_rgb)
plt.title('Ảnh Gốc (RGB)')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(hue_display, cmap='gray')
plt.title('Kênh Hue (H)')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(saturation_display, cmap='gray')
plt.title('Kênh Saturation (S)')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(value_display, cmap='gray')
plt.title('Kênh Value (V)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

4. Viết chương trình nạp một ảnh, chuyển sang hệ màu HSV. Lưu ảnh mới với kênh Hnew= 1/3 Hold, Vnew = 3/4 Vold.

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
from skimage import color

# Nạp ảnh gốc
try:
    original_image_rgb = iio.imread('bird.png')
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file 'bird.png'. Vui lòng kiểm tra lại đường dẫn.")
    exit() # Thoát chương trình nếu không tìm thấy ảnh

# Kiểm tra xem ảnh có phải là ảnh màu (có 3 RGB) hay không
if original_image_rgb.ndim != 3 or original_image_rgb.shape[2] != 3:
    print("Lỗi: Ảnh nạp vào không phải là ảnh màu RGB hoặc có số không đúng.")
    exit()

# Chuyển đổi ảnh từ hệ màu RGB sang HSV
hsv_image = color.rgb2hsv(original_image_rgb)

# Trích xuất các kênh H, S, V
hue_old = hsv_image[:, :, 0]
saturation_old = hsv_image[:, :, 1]
value_old = hsv_image[:, :, 2]

# Tính toán các kênh H_new và V_new theo yêu cầu
hue_new = hue_old / 3
value_new = value_old * 3 / 4

# Đảm bảo các giá trị vẫn nằm trong khoảng [0, 1] sau khi tính toán
hue_new = np.clip(hue_new, 0, 1)
value_new = np.clip(value_new, 0, 1)

# Kết hợp các kênh mới thành ảnh HSV mới
# Sử dụng np.stack để xếp chồng các kênh lại theo chiều mới (kênh màu)
modified_hsv_image = np.stack([hue_new, saturation_old, value_new], axis=-1)

# Chuyển đổi ảnh HSV mới trở lại hệ màu RGB
# hsv2rgb trả về ảnh float với giá trị trong khoảng [0, 1]
modified_rgb_image_float = color.hsv2rgb(modified_hsv_image)

# Chuyển đổi ảnh RGB mới về dạng uint8 [0-255] để lưu và hiển thị
modified_rgb_image_uint8 = (modified_rgb_image_float * 255).astype(np.uint8)

# Lưu ảnh mới
iio.imsave('bird_hsv_modified.png', modified_rgb_image_uint8)
print("\nĐã lưu ảnh mới: bird_hsv_modified.png")

# Hiển thị các ảnh
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(original_image_rgb)
plt.title('Ảnh Gốc (RGB)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(modified_rgb_image_uint8)
plt.title(r'Ảnh đã chỉnh HSV ($H_{new}=H_{old}/3$, $V_{new}=V_{old}*3/4$)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

5. Viết chương trình sử dụng mean filter cho các hình trong thư mục Exercise.

```python
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as sn
import matplotlib.pylab as plt
import colorsys

# opening the image and converting it to grayscale
baby = iio.imread('./exercise/baby.jpeg', mode='F')
balloons_noisy = iio.imread('./exercise/balloons_noisy.png', mode='F')
flower = iio.imread('./exercise/flower.jpeg', mode='F')

# initializing the filter of size 5 by 5
#the filter is divided by 25 for normalization
k = np.ones((5,5))/25

# performing convolution
baby_mean_filter = sn.convolve(baby, k).astype(np.uint8)
balloons_noisy_mean_filter = sn.convolve(balloons_noisy, k).astype(np.uint8)
flower_mean_filter = sn.convolve(flower, k).astype(np.uint8)

iio.imsave('./exercise/baby_mean_filter.png', baby_mean_filter)
iio.imsave('./exercise/balloons_noisy_mean_filter.png', balloons_noisy_mean_filter)
iio.imsave('./exercise/flower_mean_filter.png', flower_mean_filter)

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(baby_mean_filter)
plt.title('baby_mean_filter')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(balloons_noisy_mean_filter)
plt.title('balloons_noisy_mean_filter')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(flower_mean_filter)
plt.title('flower_mean_filter')
plt.axis('off')
plt.show()
```
